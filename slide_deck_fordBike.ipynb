## Distribution of Duration seconds

> After ploting histogram of durationtime Found after log values must furqancy time is between 600 sec and 650 sec that time of travel in most

```python
log_binsize = 0.05
plt.figure(figsize=(15,8))

bins = 10 ** np.arange(2.4, np.log10(df['duration_sec'].max())+log_binsize, log_binsize)
plt.hist(data = df, x = 'duration_sec',bins=bins);
plt.xscale('log')
plt.xlim((0,1e4))
plt.xticks([450,500,600,700,800,1000, 1500, 4000,10000], [450,500,600,700,800, '1k','1.5k','4k','10k'])

plt.xlabel('Duration Seconds')
plt.show()
```

## Box ploting Period of day and Duration seconds

> Period of day thare big outlier rang,afternoon time heighr IQR and heighr ,Min Dueation seconds is same of all period time of day

```python
plt.figure(figsize=(15,6))
sb.boxplot(data = df, x = 'period_day', y = 'duration_sec');
plt.xlabel('Period Of Day')
plt.ylabel('Duration Seconds count')
plt.ylim((0,2500))
plt.show()
```

## User Type and duration seconds

>by user type avreg duration seconds for trip more highre than subscriber users

```python
plt.figure(figsize=(15,6))
sb.boxplot(data = df, x = 'user_type', y = 'duration_sec');
plt.xlabel('User type')
plt.ylabel('Duration Seconds')
plt.ylim((0,5000))
plt.show()
```

## Month and period of day
>n plot from may to october increes trip in period afternoon,but other month allmost morning is biger than other period

```python
plt.figure(figsize=(15,6))
sb.countplot(data = df, y = 'month', hue = 'period_day',hue_order=['morning','afternoon','night'])
plt.ylabel('Month Name')
plt.title('period Day using in Months')
plt.show()
```

## Weekday and period of day
>weekday trips avreg of duration seconds afternoon period is higher than other period and highr on saturday and friday

```python
plt.figure(figsize=(12,10))
sb.boxplot(data = df, y = 'day', x = 'duration_sec', hue = 'period_day')
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.xlim((0,2500))
plt.xlabel('Duration Seconds')
plt.title('Period time of day in week day')
plt.show()
```
